```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
d={'Name':['Aruna','Bala','Chandru','Dravid','Devi','Priya'],'Age':[20,22,23,27,24,90],'Gender':['F','M','M','M','F','F'], 'Score':[20,25,30,60,50,70]}
Marks=pd.DataFrame(d)
Marks
```

```python
plt.plot(Marks["Age"], Marks["Score"], marker='o', markerfacecolor='r')
```

```python
plt.boxplot(Marks["Age"],vert=False)
plt.title('Age boxplot')
```

```python
plt.boxplot(Marks["Score"],vert=False)
plt.title('Age boxplot')
```

```python
import pandas as pd
d={'Dessert':['Cake','Chocolate','icecream','Bread cake','Velvet cake'], 'preference':[7,8,7,6,6.5]}
c_df = pd.DataFrame(d)
c_df
```

```python
plt.pie(c_df["preference"],labels=c_df['Dessert'],startangle=90,autopct="%.1f%%",radius=2,explode=[0,0,0,0,0])
centre_circle = plt.Circle((0, 0),1, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
```

```python
fig,ab=plt.subplots(2,2)
x=np.arange(1,10)
ab[0][0].plot(x,x*x)
ab[0][0].set_title('square')
ab[0][1].plot(x,np.sqrt(x))
ab[0][1].set_title('square root')
ab[1][0].plot(x,np.exp(x))
ab[1][0].set_title('exp')
ab[1][1].plot(x,np.log10(x))
ab[1][1].set_title('log')
plt.show()
```

```python
d1=plt.subplot2grid ((3,3),(0,0),colspan=2)
d2=plt.subplot2grid ((3,3),(0,2),colspan=2)
d3=plt.subplot2grid ((3,3),(1,0),rowspan=2, colspan=2)
x=np.arange(1,10)
d1.plot(x,x*x)
d1.set_title('square')
d2.plot(x,np.sqrt(x))
d2.set_title('square root')
# d3.plot(x,np.exp(x))
# d3.set_title('exp')
d3.plot(x,np.log10(x))
d3.set_title('log')
plt.show()
```
