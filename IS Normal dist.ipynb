```python
import scipy.stats as stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
a=[]
for i in range(0,1000):
    a.append(random.randint(1,5))
```

```python
a=np.array(a)
```

```python
np.size(a)
```

```python
pop_mean=np.mean(a)
pop_mean
```

```python
sample_mean=[]
for j in range(0,10):
    sample_mean.append(np.mean(random.choices(a,k=5)))
```

```python
sample_mean
```

```python
np.size(sample_mean)
```

```python
plots=sns.distplot(a) 
plots.vlines(x = np.mean(a),ymin=0, ymax=1,linestyles="solid")       # mean of poopulation
plots.vlines(x = np.mean(sample_mean),ymin=0, ymax=1,linestyles="--") # mean of sample means
```

```python
plots1=sns.distplot(sample_mean)
plots1.vlines(x = np.mean(a),ymin=0, ymax=1,linestyles="solid")  # mean of poopulation
plots1.vlines(x = np.mean(sample_mean),ymin=0, ymax=1,linestyles="--") # mean of sample means
```
